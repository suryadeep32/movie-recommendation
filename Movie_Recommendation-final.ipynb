```python
movies = movies.merge(credits,on='title')
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L = []
    counter=0
    for i in ast.literal_eval(text):
        if counter!=3:
            L.append(i['name']) 
            counter+=1
        else:
            break
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L 
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample(5)
```

```python
movies['genres'].apply(lambda x:[i.replace(" ",'')for i in x])
movies['keywords'].apply(lambda x:[i.replace(" ",'')for i in x])
movies['cast'].apply(lambda x:[i.replace(" ",'')for i in x])
movies['crew'].apply(lambda x:[i.replace(" ",'')for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
#new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
#list to string
new.head(8)
```

```python
new['tags']= new['tags'].apply(lambda x:x.lower())
#recommnended
```

```python
new.head()
```

```python
import nltk
# remove morphological affixes from words, leaving only the word stem
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
ps.stem('loving')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
#cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
#new[new['title'] == 'Batman'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    movie_lists=similarity[index]
    distances = sorted(list(enumerate(movie_lists)),reverse=True,key = lambda x: x[1])[1:6]
    for i in distances:
        print(new.iloc[i[0]].title)
        
```

```python
recommend('Tangled')
```

```python
recommend('Avatar')
```
